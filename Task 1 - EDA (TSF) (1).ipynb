- UNIVARIATE ANALYSIS

1. HISTOGRAM 

```python
#plotting histogram for each continous variable 
df.hist(figsize=(18,24));
```

2. BOXPLOT

```python
df.boxplot(figsize=(10,8))  
plt.xticks(rotation=90)
plt.show()
#int and float only
```

3. COUNTPLOT

```python
sns.countplot(x='Ship Mode', data= df)
plt.show()
```

```python
sns.countplot(x='Segment', data= df)
plt.show()
```

```python
plt.figure(figsize=(25,14))
sns.countplot(x= 'Sub-Category', data=df)
plt.show()
```

Category vs Region 

```python
sns.countplot(x= 'Category', hue='Region', data= df)
plt.show()
```

- Insights drawn :

    - Postal code shows no presence of outliers, rest all variables have outliers present in the data
    - 'Standard class' is considered to be most used ship mode 
    - Out of all the segments, 'Consumer' has the highest count
    - Sub-category -'Binders' is the highest sell product.
    - Category - 'Office Supplies'has the highest sale in all 4 regions (Most in the western region)
